### 1. Handle Missing Data

```python
# Identify which columns have missing or null values.
loans.isna().sum()
```

##### ***Insights:*** These are the columns with missing or null values: 
- Gender: **13 Missing values**
- Married: **3 Missing values**
- Dependents: **15 Missing values**
- Self_Employed: **32 Missing values**
- LoanAmount: **22 Missing values**
- Loan_Amount_Term: **14 Missing values**
- Credit_History: **50 Missing values**

```python
# Choose the appropriate strategy:
# A. Remove rows or columns with too many missing entries.
## Get the rows that have missing values in all the columns
loans[loans.isna().all(1)]
```

##### ***Insights:*** 
- There are no rows with missing values in all the columns. So, we don't need to remove any rows. 
- We don't remove any columns, because they contain important information.

```python
# B. Fill missing values using suitable methods such as:
# 1. Mean or median (for numerical values)
# 2. Mode or placeholder (for categorical values)

for i in loans.columns:
    if loans[i].isna().sum() > 0:
        print(i + " : " ,loans[i].isna().sum(), "(" + str((loans[i].dtype)) + ")")
```

##### ***Insights:*** 
- There are 7 columns with missing values.
- Gender, Married, Dependents, Self_Employed are categorical columns.
- LoanAmount, Loan_Amount_Term, Credit_History are numerical columns.

```python
# Fill missing values of categorical columns with mode
loans['Gender'] = loans['Gender'].fillna(loans['Gender'].mode()[0])
loans['Married'] = loans['Married'].fillna(loans['Married'].mode()[0])
loans['Dependents'] = loans['Dependents'].fillna(loans['Dependents'].mode()[0])
loans['Self_Employed'] = loans['Self_Employed'].fillna(loans['Self_Employed'].mode()[0])
loans.isna().sum()
```

##### ***Insights:*** We replaced missing values of all categorical columns with **mode**.

```python
# Fill missing values of numerical columns
# LoanAmount column
loans['LoanAmount'].unique()
```

##### ***Insights:*** By checking all the unique values of LoanAmount column, we can replace missing values of LoanAmount with median.

```python
# Fill missing values with median
loans['LoanAmount'] = loans['LoanAmount'].fillna(loans['LoanAmount'].median())
loans['LoanAmount'].isna().sum()
```

##### ***Insights:*** We replaced missing values of LoanAmount column with **median**.

```python
# Fill missing values of numerical columns
# Loan_Amount_Term column
loans['Loan_Amount_Term'].unique()
```

```python
loans['Loan_Amount_Term'].value_counts()
```

```python
# We can fill missing values with mode
loans['Loan_Amount_Term'] = loans['Loan_Amount_Term'].fillna(loans['Loan_Amount_Term'].mode()[0])
loans['LoanAmount'].isna().sum()
```

##### ***Insights:*** By checking all the unique values and value counts of Loan_Amount_Term column, we replaced missing values of Loan_Amount_Term with mode or 360.0, which is the most common value.

```python
# Fill missing values of numerical columns
# Credit_History column
loans['Credit_History'].unique()
```

```python
loans['Credit_History'].value_counts()
```

```python
# We can fill missing values with mode
loans['Credit_History'] = loans['Credit_History'].fillna(loans['Credit_History'].mode()[0])
loans['Credit_History'].isna().sum()
```

##### ***Insights:*** By checking all the unique values and value counts of Credit_History column, we replaced missing values of Credit_History with mode or 1.0, which is the most common value.

```python
loans.isna().sum()
```

##### ***Insights:*** There are no missing values in the dataset now.

##### **Justify the method chosen based on the type of data and its significance:**
- For categorical data, we replaced missing values with **mode**.
- For *LoanAmount* we replaced missing values with **median**, cause it is a **continuous numerical** column.
- For *Loan_Amount_Term* and *Credit_History* we replaced missing values with **mode**, cause they are **discrete numerical** columns.
